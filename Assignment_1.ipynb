# Ques 1 Write a program which will find all such numbers which are divisible by 7 but are not a multipleof 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
num=[]
for i in range(2000,3201):
    if i%7==0 and i%5!=0:
        num.append(i)
print(num)        
```

# Ques 2 Create the below pattern using nested for loop in Python.

```python
m=5
for i in range(m):
    for j in range(i):
        print("*",end=" ")
    print('')
for k in range(m,0,-1):
    for l in range(k):
        print("*",end=" ")
    print('')    
           
    
```
